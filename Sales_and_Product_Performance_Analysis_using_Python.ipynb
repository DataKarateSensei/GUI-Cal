```python
import pandas as pd
import numpy as np
import string
import matplotlib
import matplotlib.pyplot as plt
import random
import seaborn as sns
```

```python
products=["Product A","Product B","Product C","Product D","Product E"]
Area=["Colombo","Jaffna","Kandy","Galle","Anuradhapura"]
date=pd.date_range('2022-01-01', '2022-01-31', freq='D')
```

```python
#creating probable values for columns in the data frame
def generate_random_prodID():
  return random.choices(products,weights=[0.15,0.32,0.28,0.20,0.05])
def generate_random_sales_quantity():
  return int(random.uniform(50,100))
def generate_random_date():
  return random.choice(date)
def generate_random_area():
  return random.choice(Area)
```

```python
#Creating the data frame for 100 records
total_number_of_transactions=100
data=[]
for i in range(total_number_of_transactions):
  row=[]
  row.append(generate_random_prodID())
  row.append(generate_random_sales_quantity())
  row.append(generate_random_date())
  row.append(generate_random_area())
  data.append(row)
df=pd.DataFrame(data,columns=["Product ID","Sales Quan","Transaction Date","Sales Area"])
df
```

```python
#Creating a Pie Chart for Sales Quantity by Area
sales_quan_by_area=df.groupby(["Sales Area"]).sum()
sales_quan_by_area
```

```python
plt.pie(sales_quan_by_area["Sales Quan"],labels=sales_quan_by_area.index)
plt.title("Sales by Region")
plt.show()
```

```python
home=[20, 34, 30, 35, 27]
elec=[25, 32, 34, 20, 25]
```

```python
#creating a grouped bar chart for Sales by Region
x=np.arange(len(Area))
width=0.35

fig, ax=plt.subplots()
rects1 = ax.bar(x - width/2, home, width, label='Home Supplies')
rects2 = ax.bar(x + width/2, elec, width, label='Electronics')

ax.set_ylabel("Sales")
ax.set_title("Sales for Home Suppies & Electronics by Region")
ax.set_xticks(x)
ax.set_xticklabels(Area)
ax.legend()

fig.tight_layout()

plt.show()
```
